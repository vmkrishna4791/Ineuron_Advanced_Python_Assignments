```python
Q4. What are bound methodsÂ and how do we use them?

"""
(A)A bound method is the one which is dependent on the instance of the class as the first argument. 
It passes the instance as the first argument which is used to access the variables and functions.

e.g.

class A:
  
    def func(self, arg):
        self.arg = arg
        print("Value of arg = ", arg)
  
  
# Creating an instance
obj = A()  
  
# bound method
print(obj.func)

The instance obj is automatically passed as the first argument to the function 
called and hence the first parameter of the function will be used to access the 
variables/functions of the object.



"""



```

```python
#Q5. What is the purpose of pseudoprivate attributes?
"""
(A)Pseudoprivate attributes are also useful in larger frameworks or tools, both to avoid introducing new method names
that might accidentally hide definitions elsewhere in the class tree and to reduce the chance of internal methods being 
replaced by names defined lower in the tree. If a method is intended for use only within a class that may be mixed 
into other classes, the double underscore prefix ensures that the method won't interfere with other names in the tree, 
especially in multiple-inheritance

"""

#e.g

class C1:    
    def method1(self):
        self.X = 25 
        
    def method2(self):
        print (self.X)
        
        
class C2:
    def methoda(self): 
        self.X = 45             
        
    def methodb(self): 
        print (self.X)
        
class C3(C1, C2):
    
    def __init__(self):
        pass

    
I=C3()

I.method1()
I.method2()
I.methodb()

I.methoda()
I.methodb()
I.method2()


#####Using the pseudoprivate attributes ######

class C1:     
    
    def method1(self): 
        self.__X = 88      
    
    def method2(self): 
        print (self.__X)      # Becomes _C1__X in I 

class C2:    
    
    def methoda(self): 
        self.__X = 99           
        
    def methodb(self): 
        print (self.__X)      # Becomes _C2__X in I 

class C3(C1, C2): 
    pass 
    
    
I = C3()

#I.method1()
#print(I.method2())

#print(I.methodb())

I.methoda()
print(I.methodb())
print(I.method2())
```
